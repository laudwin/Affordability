```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
# Load dataset
file_path = "model_data.csv"
df = pd.read_csv(file_path)
```

```python
# Display basic info
display(df.info())
display(df.describe())
```

```python
def plot_distribution(column):
    plt.figure(figsize=(8, 4))
    sns.histplot(df[column], bins=30, kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
# Plot distributions of numerical features
num_features = ['age', 'avg_call_duration', 'num_calls_per_day', 'num_sms_per_day', 'data_usage_mb', 'monthly_topups', 'avg_topup_amount', 'geographic_mobility', 'default_risk_score']
for feature in num_features:
    plot_distribution(feature)
```

```python
# Encode categorical variables
df = pd.get_dummies(df, columns=['gender', 'income_category'], drop_first=True)
```
